| Technique                          | Description                                                                                         | Pros                                                                 | Cons                                                                 |
|------------------------------------|-----------------------------------------------------------------------------------------------------|----------------------------------------------------------------------|----------------------------------------------------------------------|
| **Mean Imputation**                | Replace with mean.                                                                                | Simple, fast.                                                        | Underestimates variance, not robust to outliers.                     |
| **Median Imputation**              | Replace with median.                                                                              | Robust to outliers, skewed data.                                     | Less sensitive to distribution.                                      |
| **K-Nearest Neighbors (KNN)**      | Use KNN to find neighbors and impute with their average.                                          | Captures local patterns.                                             | Computationally intensive.                                           |
| **MICE**                           | Iteratively model each feature as a function of others.                                             | Handles complex relationships.                                         | More complex, computationally intensive.                               |
| **Regression Imputation**          | Use regression to predict missing values.                                                         | Captures complex relationships.                                        | May overfit.                                                         |
| **Interpolation**                  | Interpolate between known values.                                                                   | Useful for time series.                                              | Assumes linear relationships.                                          |
| **Bayesian Methods**               | Use Bayesian inference to estimate missing values.                                                | Incorporates prior knowledge.                                        | Complex to implement, requires expertise.                              |
| **Deep Learning**                  | Use neural networks to learn data distribution and impute.                                        | Captures complex patterns.                                           | Requires significant data and resources.                               |
| **Random Value Imputation**        | Replace with random values from the column's distribution.                                        | Simple and quick.                                                    | Introduces noise, can bias the model, reduces data quality.          |

```python
import pandas as pd
import numpy as np
```

```python
data = {'A':[1,2,np.nan,4,5,np.nan,7]}
df = pd.DataFrame(data)
```

```python
mean_value = df['A'].mean()
std_value = df['A'].std()
```

```python
random_values = np.random.normal(mean_value, std_value, df['A'].isna().sum())
```

```python
# Fill missing values with random values
df['A_random_imputed'] = df['A'].copy()
df.loc[df['A'].isna(), 'A_random_imputed'] = random_values

print("\nRandom Value Imputation:")
print(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Sample DataFrame
data = {'A': [1, 2, np.nan, 4, 5, np.nan, 7]}
df = pd.DataFrame(data)

# Median Imputation
imputer = SimpleImputer(strategy='median')
df['A_median_imputed'] = imputer.fit_transform(df[['A']])

print("\nMedian Imputation:")
print(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Sample DataFrame
data = {'A': [1, 2, np.nan, 4, 5, np.nan, 7]}
df = pd.DataFrame(data)

# Mean Imputation
imputer = SimpleImputer(strategy='mean')
df['A_mean_imputed'] = imputer.fit_transform(df[['A']])

print("Mean Imputation:")
print(df)
```
