```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("train.csv")
```

```python
print("Before")
print(df[['Age']].head())
```

```python
def normalize_age(X):
  max_age = X.max()
  return X/max_age
```

```python
age_transformer = FunctionTransformer(normalize_age)
```

```python
df['Age_normalized'] = age_transformer.fit_transform(df[['Age']])
```

```python
print("\nAfter Transformation:")
print(df[['Age', 'Age_normalized']].head())
```
