```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python

df.sample(5)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Step 2: Define the columns to encode
categorical_columns = ['gender','review','education','purchased']
```

```python
# Step 3: Initialize the OrdinalEncoder
# Define the categories for each column in the correct order
encoder = OrdinalEncoder(
    categories=[
        ['Male', 'Female'],  # gender: Male=0, Female=1
        ['Poor', 'Average', 'Good'],  # review: Poor=0, Average=1, Good=2
        ['School', 'UG', 'PG'],  # education: School=0, UG=1, PG=2
        ['No', 'Yes']  # purchased: No=0, Yes=1
    ]
)
```

```python
# Step 4: Apply ordinal encoding to the categorical columns
df[categorical_columns] = encoder.fit_transform(df[categorical_columns])
```

```python
df
```

Assigns integers to categories based on a predefined order or ranking.

Use Case : Suitable for categorical variables with an inherent order (e.g., "Low," "Medium," "High").

Implementation : Can be done manually using mappings or using OrdinalEncoder from sklearn.preprocessing.

Behavior :
Encodes categories based on their rank or order.

Preserves the ordinal relationship between categories.

##Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('customer.csv')
```

```python
le = LabelEncoder()
```

```python
data['gender'] = le.fit_transform(data['gender'])
data['review'] = le.fit_transform(data['review'])
data['education'] = le.fit_transform(data['education'])
data['purchased'] = le.fit_transform(data['purchased'])
```

```python
data
```

##Applying Encoding :

Each categorical column (gender, review, education, purchased) is encoded using the fit_transform() method.

LabelEncoder assigns a unique integer to each category ***alphabetically*** :

For gender: "Female"  0, "Male"  1

For review: "Average"  0, "Good"  1, "Poor"  2

For education: "PG"  0, "School"  1, "UG"  2

For purchased: "No"  0, "Yes"  1

##Difference
Label Encoding assigns unique integers to categories alphabetically (no order assumed),
while Ordinal Encoding assigns integers based on a predefined order or ranking (order is assumed).
