```python
class SimpleLinearRegression:
    def __init__(self):
        # Initialize slope (m) and intercept (b) to None
        self.m = None  # Slope of the line
        self.b = None  # Intercept of the line

    def fit(self, X_train, y_train):
        """
        Train the model using training data.
        X_train: List or array of input values (independent variable).
        y_train: List or array of output values (dependent variable).
        """
        # Calculate the mean of X_train and y_train
        mean_X = sum(X_train) / len(X_train)
        mean_y = sum(y_train) / len(y_train)

        # Initialize numerator and denominator for slope calculation
        numerator = 0
        denominator = 0

        # Loop through each data point
        for i in range(len(X_train)):
            # Update numerator and denominator
            numerator += (X_train[i] - mean_X) * (y_train[i] - mean_y)
            denominator += (X_train[i] - mean_X) ** 2

        # Calculate slope (m) and intercept (b)
        self.m = numerator / denominator
        self.b = mean_y - (self.m * mean_X)

        # Print the results for clarity
        print(f"Slope (m): {self.m}")
        print(f"Intercept (b): {self.b}")

    def predict(self, X_test):
        """
        Predict the output for new input values.
        X_test: List or array of new input values.
        Returns: List of predicted output values.
        """
        # Use the equation of a line: y = m * x + b
        predictions = [self.m * x + self.b for x in X_test]
        return predictions
```

```python
# Training data
X_train = [1, 2, 3, 4, 5]  # Input values
y_train = [2, 4, 5, 4, 5]  # Output values

# Create an instance of the model
model = SimpleLinearRegression()

# Train the model
model.fit(X_train, y_train)

# Test data
X_test = [6, 7, 8]  # New input values

# Make predictions
predictions = model.predict(X_test)
print("Predictions:", predictions)
```
