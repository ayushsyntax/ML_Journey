```python
import numpy as np
import pandas as pd

data = {
    'Feature1': [10, 20, np.nan, 40, 50],
    'Feature2': [1.1, 2.2, 3.3, np.nan, 5.5],
    'Feature3': [100, 200, 300, 400, np.nan]
}
df = pd.DataFrame(data)
print("Original Dataset:")
print(df)
```

```python
from sklearn.impute import SimpleImputer

# Replace missing values with the mean of each column
simple_imputer = SimpleImputer(strategy='mean')
simple_imputed_data = simple_imputer.fit_transform(df)

simple_imputed_df = pd.DataFrame(simple_imputed_data, columns=df.columns)
print("\nDataset After SimpleImputer:")
print(simple_imputed_df)
```

```python
from sklearn.impute import KNNImputer

# Replace missing values using 2 nearest neighbors
knn_imputer = KNNImputer(n_neighbors=2,weights='uniform')
knn_imputed_data = knn_imputer.fit_transform(df)

knn_imputed_df = pd.DataFrame(knn_imputed_data, columns=df.columns)
print("\nDataset After KNNImputer:")
print(knn_imputed_df)
```
